```julia
using CSV, DataFrames, Plots
```

```julia
file = "/home/arturofburgos/OpenFoamTutorials/MixingTank/Test2/postProcessing/spargerGasProbes/0/alpha.air"

df = CSV.read(file, DataFrame; comment="#", delim=' ', ignorerepeated=true,
              header=["Time","Probe0","Probe1","Probe2","Probe3"])

t = df.Time                              # eixo X
Y = Matrix(df[:, Not(:Time)])            # todas as colunas‑sonda numa matriz

plot(t, Y;                               # x = vetor, y = matriz (uma curva por coluna)
     labels = names(df)[2:end],          # etiquetas de cada sonda
     xlabel = "Time [s]",
     ylabel = "α_air",
     title  = "Probes – α_air")
```
